# Challenge 3 - Constructing Confidence Intervals

We will test whether the hourly wage of all hourly workers is significantly different from $30/hr.

In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. Is $30/hr within that interval?

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
#
#calculate the mean
mean = salaries['Hourly Rate'].mean()
#calculate the degrees of freedom
degrees_of_freedom =   salaries['Hourly Rate'].count() - 1
#calculate the standard error
standard_error = sem(salaries['Hourly Rate'])
#calculate the confidence interval and print it
confidence_interval = t.interval(confidence=0.95, df=degrees_of_freedom, loc=mean, scale=standard_error)
print(confidence_interval)
```

This is fine if we have thousands of worker data. But what if we have only 100 workers data?

Sample 100 workers and re-construct the 95% confidence interval. Is the interval wider of narrower? And why?
Do you still encapsulate the $30/hr mark in this case?

```python
# Create a sample of 100 salaries

salaries_sample = salaries.sample(100)

# Calculate the sample mean and standard deviation
sample_mean = salaries_sample['Hourly Rate'].mean()


sample_standard_error = sem(salaries_sample['Hourly Rate'])

sample_degrees_of_freedom = salaries_sample['Hourly Rate'].count() - 1

# Calculate the confidence interval
confidence_interval = t.interval(0.95, df=sample_degrees_of_freedom, loc=sample_mean, scale=sample_standard_error)

print(confidence_interval)

# The confidence interval widens quite a bit due to the smaller sample size.  The 30 hr mark is not encapsulated here either. 
```
